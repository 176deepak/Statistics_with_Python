# Probability Distributions Plots
- Discreate Distributions
- Continuous Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import seaborn as sns
```

## Discrete Distributions
- Bernouli Distribution
- Binomial Distribution
- Uniform Distribution
- Poision Distribution

### Bernouli Distribution

```python
# probability of success, means getting 1
p = 0.5  

# probability of failure, means getting 0 
q = 1-p

# creating bernouli random variable 
bernouli = sp.stats.bernoulli(p) 

# generating 100 random variates
rv = bernouli.rvs(100)              

# plotting distribution
unique, counts = np.unique(np.array(rv), return_counts=True)

# bar graph for bernouli distribution
plt.bar(unique, counts, tick_label = [0, 1], color='rebeccapurple')

# plot details
plt.title(f'Bernoulli Distribution (p={p})')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.grid(axis='both', linestyle='--', alpha=1)
plt.show()
```

### Binomial Distribution

```python
# number of trials
n = 50

# probability of getting success
p = 0.4

# probability of getting failure
q = 1-p

# generating random samples that follow binomial distribution
data = np.random.binomial(n, p, 1000)

# distribution plotting
plt.figure(figsize=(20, 10))
density, bins, _ = plt.hist(data, density=True, alpha = 0.75, color='blue')
bin_centers = (bins[1:] + bins[:-1]) / 2
plt.plot(bin_centers, density, color='r', linestyle='-', marker='o')
# plot details
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(range(n + 1))
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

### Uniform Distribution

```python
# Lower bound of the interval
a = 0  

# Upper bound of the interval
b = 10  

# Number of data points to generate
size = 1000  

# Generate a set of data points following a uniform distribution
x = sp.stats.uniform.rvs(0.01,0.99,size=1000)

# ploting distribution
plt.hist(x, density = True, color='b')
plt.axhline(y=sp.stats.uniform.pdf(x[0]),color='r')
plt.title('Uniform distribution')
plt.ylabel('Density')
plt.xlabel('X')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

### Poision Distribution

```python
# parameter of poision distribution
_lambda1 = 1

# generating random samples
data1 = np.random.poisson(_lambda1, 1000)

# plotting distribution
plt.hist(data1, density=True, alpha = 0.75, color='blue')
plt.title(f'Poisson Distribution (λ={_lambda1})')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Continuous Distribution
- Normal/Gaussian Distribution
- Exponential Distribution
- Uniform/Rectangular Distribution
- t Distribution
- Chi-Squared Distribution

### Normal Distribution

```python
# Mean
mu = 0

# Standard deviation
sigma = 1  

# Number of data points to generate
size = 1000  

# Generating a set of data points following a Gaussian distribution
data = np.random.normal(mu, sigma, size)

# Create a histogram to visualize the distribution
density, bins, _ = plt.hist(data, bins=20, density=True, alpha=0.6, color='blue')
bin_centers = (bins[1:] + bins[:-1]) / 2
plt.plot(bin_centers, density, color='r', linestyle='-')
plt.title(f'Gaussian Distribution (μ={mu}, σ={sigma})')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Exponential Distribution

```python
# You can adjust this parameter
lambda_ = 0.5  

# Generate a set of data points following an exponential distribution
data = np.random.exponential(scale=1/lambda_, size=1000)

# Create a histogram to visualize the distribution
density, bins, _ = plt.hist(data, bins=20, density=True, alpha=0.6, color='blue')
bin_centers = (bins[1:] + bins[:-1]) / 2
plt.plot(bin_centers, density, color='r', linestyle='-')
plt.title(f'Exponential Distribution (λ={lambda_})')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Uniform Distribution

```python
# Lower bound of the interval
a = 0  

# Upper bound of the interval
b = 10  

# Number of data points to generate
size = 1000  

# Generate a set of data points following a continuous uniform distribution
data = np.random.uniform(a, b, size)

# Create a histogram to visualize the distribution
plt.hist(data, bins=20, density=True, alpha=0.6, color='blue')
plt.title(f'Continuous Uniform Distribution [{a}, {b}]')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### t Distribution

```python
# Parameters for the t-distribution
df = 10  # Degrees of freedom (adjust this as needed)

# Generate data points to represent the t-distribution
x = np.linspace(sp.stats.t.ppf(0.001, df), sp.stats.t.ppf(0.999, df), 1000)
y = sp.stats.t.pdf(x, df)  # Probability density function

# Create a line plot to visualize the t-distribution
plt.plot(x, y, 'r-', lw=2, alpha=0.6, label='t-distribution PDF')
plt.title(f't-Distribution (Degrees of Freedom = {df})')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(linestyle='--', alpha=0.7)
plt.legend(loc='best')
plt.show()
```

### Chi-Squared Distribution

```python
# Parameters for the chi-squared distribution
df = 5 

# Generate data points to represent the chi-squared distribution
x = np.linspace(0, 20, 1000)  # Adjust the range as needed
y = sp.stats.chi2.pdf(x, df)  # Probability density function

# Create a line plot to visualize the chi-squared distribution
plt.plot(x, y, 'r-', lw=2, alpha=0.6, label=f'Chi-squared PDF (df={df})')
plt.title(f'Chi-squared Distribution (Degrees of Freedom = {df})')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(linestyle='--', alpha=0.7)
plt.legend(loc='best')
plt.show()
```
